```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import pandas as pd
from scipy import interpolate
import math
df = pd.read_csv('bending_moduli_lipids.dat', delim_whitespace=True, names= ["Lipid", "Value", "SE"] )
```

```python
fig = plt.figure(1034, figsize=(14,10))
ax = fig.add_subplot(111)
Lipid = df['Lipid']
STD = df['SE'] 

#Value = df['Value']
# Read value for 2 decimal point
Value = round(df['Value'], 2)

# This is the location for the annotated text
i = 0.85
j = 2
# Annotating the bar plot with the values in column(Value)
for i in range(len(Lipid)):
    #plt.annotate(Value[i], (-0.1 + i, Value[i] + j))
    plt.annotate(Value[i], (-0.1+i,Value[i] + j),rotation=90, color='blue', fontsize=40, fontname='arial')
#thick axes
from pylab import *
rc('axes', linewidth=3)    

plt.xticks(rotation=90, horizontalalignment="center", fontsize=42, fontname='arial')
plt.ylim([0, 30.001])
plt.yticks(np.arange(0, 30.001, 5 ), fontsize=42, fontname='serif')
#plt.title("Membrane bending moduli", fontsize=24, fontname='arial')
#plt.ylabel("κ[ kT]ReSIS",loc='center',fontsize=38, fontname='arial')
plt.ylabel('$\kappa_c$ [ kT]ReSIS',loc='center',fontsize=42, fontname='arial')
plt.tick_params(axis ='both', which= 'major', labelsize=42, direction='in', length=16, top=1, right=1, width=1.7)
#plt.bar(Lipid, Value, yerr=STD, width= 0.3, alpha=0.5, align='center',color='blue', edgecolor = 'black', capsize=10)
plt.bar(Lipid, Value, yerr=STD, width= 0.3, align='center',color='blue', edgecolor = 'black', capsize=8)
# Saving the plot as a 'png'
#plt.savefig('Barplot_new.pdf',bbox_inches='tight', dpi=600)
plt.savefig('BarPlot_lipid_new.png', bbox_inches='tight', dpi=600)
# Displaying the bar plot
plt.show()
```
